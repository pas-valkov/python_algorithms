```python
from bs4 import BeautifulSoup
import pandas as pd
import os
import re
import sys

folder_name= 'content_full'

def get_doc_names(path):
    _, _, filenames = next(os.walk(path))
    return filenames

def extract_doc_info(doc_name):
    doc_info = {
        'doc_id': int(doc_name.replace('.dat', '')),
        'text': None,
    }
    try:
        global folder_name
        full_doc_name = os.path.join(folder_name, doc_name)


        with open(full_doc_name, 'r', encoding='utf-8', errors='ignore') as doc:
            url = doc.readline()
#             doc_info['url'] = url
            soup = BeautifulSoup(doc, 'html.parser')

            title = soup.find('title')
            if title is not None:
                soup.find('title').replace_with('') 

            metas = soup.find_all('meta')
            for meta in metas:
                meta.replace_with('')
                
            for noscript in soup.find_all('noscript'):
                noscript.replace_with('')

            for script in soup.find_all('script'):
                script.replace_with('')

            for style in soup.find_all('style'):
                style.replace_with('')

            doc_info['text'] = soup.text.replace('\n', ' ').replace('\t', '')
            doc_info['text'] = re.sub("\s+", ' ', doc_info['text'])


    except:
            print('\n', full_doc_name)
            exc_type, exc_value, traceback = sys.exc_info()
            print(exc_type, exc_value, traceback)
    return doc_info

doc_names = get_doc_names(folder_name)
doc_names = doc_names[:]
# doc_names = ['8569.dat', '2073.dat', '5120.dat']
```

```python
try:
    for start in range(28000,29000,1000):
        %time print(start)
        result = []
        for i, doc_name in enumerate(doc_names[start:start+1000]):
            result.append(extract_doc_info(doc_name))
            if i%10 == 0:
                print(f"\r{i} objects are processed...", end='', flush=True)

        df = pd.DataFrame(result)
        # df.sort_values(by=['doc_id'], inplace=True)

        with open('text\parsing_output_text_'+str(start)+'.csv', mode='w', encoding='utf-8', errors='ignore') as f_csv:
            df.to_csv(f_csv, index=False)
except:
        print('\n', start, i)
        exc_type, exc_value, traceback = sys.exc_info()
        print(exc_type, exc_value, traceback)
```
